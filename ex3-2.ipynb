```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
R = 2
alpha = np.pi/4
x = R*np.cos(alpha)
y = R*np.sin(alpha)
print(x, y)
```

```python
alpha = np.linspace(0, 2*np.pi, 200)
r0 = 2
r = np.full(200, r0)
plt.polar(alpha, r)
plt.title(f'Окружность, r0 = {r0}')
plt.show()
```

```python
alpha = np.linspace(0, np.pi/4, 2)
r = (0, 1)
plt.polar(alpha, r)
plt.title('Отрезок прямой y=x в полярных координатах')
plt.show()
```
