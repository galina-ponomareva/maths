```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-2*np.pi, 2*np.pi, 201)
plt.plot(x, np.cos(x-np.pi/4)-1, label = 'k = 1, a = pi/4, b = -1')
plt.plot(x, 2*np.cos(x-np.pi), label = 'k = 2, a = pi, b = 0')
plt.plot(x, 0.5*np.cos(x+np.pi/2)+0.5, label = 'k = 0.5, a = -pi/2, b = 0.5')
plt.title(u'Графики y(x) = k∙cos(x – a) + b')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
